```python
import pandas as pd
import numpy as np
```

```python
diccionario = {
    'A':['A0','A1','A2','A3'], 
    'B':['B0','B1','B2','B3'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']
}
diccionario
```

```python
df1 = pd.DataFrame(diccionario)
df1
```

```python
df2 = pd.DataFrame({
     'A':['A4','A5','A6','A7'], 
    'B':['B4','B5','B6','B7'],
    'C':['C4','C5','C6','C7'],
    'D':['D4','D5','D6','D7']
})
df2
```

# **Concat**

```python
pd.concat([df1, df2])
#En este caso los indices heredan el indice anterior por lo que estan repetidos
```

```python
#para corregir los indices repetidos
pd.concat([df1, df2],ignore_index=True)
```

```python
#Por default lo hace por axis 0 (filas) pero puedo especificar que sea por columnas 1
pd.concat([df1, df2],ignore_index=True, axis=1)
```

# **Merge**

```python
dic_2 = {
    'key': ['k0','k1','k2','k3'],
    'A':['A0','A1','A2','A3'], 
    'B':['B0','B1','B2','B3'],

}

dic_3 = {
    'key': ['k0','k1','k2','k3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
}
```

```python
izq = pd.DataFrame(dic_2)
izq
```

```python
der = pd.DataFrame(dic_3)
der
```

```python
#merge, en este caso se hara una combinación de izquierda con derecha
#es bueno especificar que el merge se haga sobre una columna especifica
#en este caso la key
izq.merge(der, on='key')
```

```python
dic_2 = {
    'key': ['k0','k1','k2','k3'],
    'A':['A0','A1','A2','A3'], 
    'B':['B0','B1','B2','B3'],

}

dic_3 = {
    'key_2': ['k0','k1','k2','k3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
}
```

```python
izq = pd.DataFrame(dic_2)
izq
```

```python
der = pd.DataFrame(dic_3)
der
```

```python
#merge
izq.merge(der, left_on='key', right_on='key_2')
```

```python
#modificando los dataframes
dic_2 = {
    'key': ['k0','k1','k2','k3'],
    'A':['A0','A1','A2','A3'], 
    'B':['B0','B1','B2','B3'],

}

dic_3 = {
    'key_2': ['k0','k1','k2',np.nan],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
}
```

```python
izq = pd.DataFrame(dic_2)
der = pd.DataFrame(dic_3)
```

```python
izq
```

```python
der
```

```python
izq.merge(der, left_on='key', right_on='key_2')#inner join solo trae las coincidencias
```

```python
#especificar que sea un left join
izq.merge(der, left_on='key', right_on='key_2', how='left')
```

```python
#inner join
izq.merge(der, left_on='key', right_on='key_2', how='inner')
```

```python
#right join
izq.merge(der, left_on='key', right_on='key_2', how='right')
```

# **Join- Index match**

```python
#va a buscar el match a traves de indices y no de columnas como se mostró anteriormente CON MERGE.
izq = pd.DataFrame({
     'A':['A0','A1','A2'], 
     'B':['B0','B1','B2'],
}, index=['k0', 'k1', 'k2' ])
```

```python
der = pd.DataFrame({
     'C':['C0','C1','C2'],
     'D':['D0','D1','D2']
}, index=['k0', 'k2', 'k3' ])
```

```python
#ejecutando el Join que funciona muy similar al merge
izq.join(der)#Es un join hecho directamente hacia los índices
```

```python
#inner join usando el párametro how
izq.join(der, how='inner')
```

```python
izq.join(der, how='left')
```

```python
izq.join(der, how='right')
```

```python
izq.join(der, how='outer')
```
